```python
df = pd.DataFrame(df.values, columns=[column.strip() for column in df.columns])
df["Country"] = df["Country"].str.strip()
df.drop(['AMA exchange rate', 'IMF based exchange rate','Population','Currency', 'Per capita GNI', 
         'Changes in inventories', 'Final consumption expenditure', 
         'General government final consumption expenditure','Gross capital formation', 
         'Gross fixed capital formation (including Acquisitions less disposals of valuables)', 
         'Household consumption expenditure (including Non-profit institutions serving households)', 
         'Other Activities (ISIC J-P)','Total Value Added']
, axis=1, inplace=True)

# Chossing countries in the South East Asia
selected_countries = ["Indonesia"]
```

```python
countries = df[(df["Country"].isin(selected_countries))]
countries
```

```python
countries = df[(df["Country"].isin(selected_countries))]

# Filter the data to only include data from 2016 to 2021
countries = countries[(countries["Year"] >= 2016) & (countries["Year"] <= 2021)]

sns.set_style("whitegrid")
sns.set(font_scale=1)

ax = sns.lineplot(x="Year", y="Gross Domestic Product (GDP)", data=countries, hue="Country",linewidth=1, palette="hls",).set(title="GDP in Indonesia from 2016 to 2021")

plt.show()
```

```python
# Filter data for Indonesia and the desired years
indonesia_data = df[(df['Country'] == 'Indonesia') & (df['Year'] >= 2016) & (df['Year'] <= 2021)]

# Select the relevant columns for the economic indicators
economic_indicators = [
    'Agriculture, hunting, forestry, fishing (ISIC A-B)',
    'Construction (ISIC F)',
    'Exports of goods and services',
    'Imports of goods and services',
    'Manufacturing (ISIC D)',
    'Mining, Manufacturing, Utilities (ISIC C-E)',
    'Transport, storage and communication (ISIC I)',
    'Wholesale, retail trade, restaurants and hotels (ISIC G-H)'
]
```

```python
# Set the figure size
plt.figure(figsize=(12, 6))

# Plot each economic indicator
for indicator in economic_indicators:
    plt.plot(indonesia_data['Year'], indonesia_data[indicator], label=indicator)

# Set the title and labels
plt.title('Fluctuations of Economic Indicators in Indonesia')
plt.xlabel('Year')
plt.ylabel('Value')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Calculate the growth rates for each economic indicator
growth_rates = indonesia_data[economic_indicators].pct_change() * 100

# Set the figure size
plt.figure(figsize=(12, 8))

# Plot the growth rates for each economic indicator
for indicator in economic_indicators:
    plt.plot(indonesia_data['Year'], growth_rates[indicator], label=indicator)

# Set the title and labels
plt.title('Growth Rates of Economic Indicators in Indonesia (2016-2021)')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
